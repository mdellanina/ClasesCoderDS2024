```python
mean=exercise.pulse[exercise.kind=="rest"].mean()
suma=exercise.pulse[exercise.kind=="rest"].sum()
cant= len(exercise.pulse[exercise.kind=="rest"])
print (suma/cant)
print (mean)
```

```python
df=exercise
pulse_low_fat= df[(df['diet'] == 'low fat') & (df['kind'] == 'rest')]
plot1=df[(df['kind']=="rest")&(df['diet']=="low fat")]
print(plot1)
sns.histplot(data=plot1, x="pulse", binwidth=5)
# Configurar etiquetas y ttulo del grfico
plt.xlabel('Frecuencia Cardaca')
plt.ylabel('Cantidad de Personas')
plt.title('Histograma de Frecuencia Cardaca de Personas en Reposo (Dieta "Low Fat")')
plt.show()
plot2=df[(df['diet'] == 'low fat') & (df['kind'] == 'running') & (df['time'] == '30 min')]
sns.histplot(data=plot2, x="pulse", binwidth=5)
plt.xlabel('Frecuencia Cardaca')
plt.ylabel('Cantidad de Personas')
plt.title('Histograma de Frecuencia Cardaca de Personas Corriendo 30min (Dieta "Low Fat")')
plt.show()
```
